## Part 2: While Loops for Validation (15 minutes)

### 🛡️ Smart Input Validation

Your finance tracker needs to handle invalid inputs gracefully. Let's start with the most important validation pattern:

````python
# Example: Keep asking until valid dollar amount
print("=== Expense Entry with Validation ===")

# We'll ask for an expense amount and keep asking until it's valid
expense = -1  # Start with invalid value

while expense < 0:  # Keep going while invalid
    try:
        expense_str = input("Enter expense amount: $")
        expense = float(expense_str)

        if expense < 0:
            print("Expense cannot be negative! Try again.")
    except ValueError:
        print("Please enter a valid number!")
        expense = -1  # Keep loop going

print(f"Valid expense recorded: ${expense:.2f}")
````

````python
# ===== YOUR CHALLENGE: BUDGET VALIDATION =====
print("=== Monthly Budget Entry ===")

budget = 0

while budget < 100 or budget > 50000:
    try:
        budget_str = input("Enter monthly budget ($100-$50,000): $")
        budget = float(budget_str)

        if budget < 100:
            print("CELESTI says: Budget too low! Minimum $100")
        elif budget > 50000:
            print("CELESTI says: Budget too high! Maximum $50,000")
    except ValueError:
        print("CELESTI says: Please enter a valid number!")
        budget = 0

print(f"Budget set to: ${budget:.2f}")
````

## Part 3: Sentinel Values for Unknown Repetition (10 minutes)

### 🏁 The "Done" Pattern

Sometimes you don't know how many times to repeat. You need a **sentinel value** - a special input that means "stop".

````python
# Example: Collect expenses until user says "done"
print("=== Multiple Expense Entry ===")

expenses = []  # List to store all expenses
total = 0      # Running total

print("Enter your expenses (type 'done' when finished):")

while True:  # Infinite loop - we'll break out when ready
    expense_input = input("Expense amount (or 'done'): $")

    # Check if user wants to stop
    if expense_input.lower() == 'done':
        break  # Exit the loop

    # Try to convert to number
    try:
        expense = float(expense_input)
        if expense >= 0:
            expenses.append(expense)
            total += expense
            print(f"  Added ${expense:.2f} | Running total: ${total:.2f}")
        else:
            print("  Expense must be positive!")
    except ValueError:
        print("  Please enter a valid number or 'done'")

print(f"\n📊 Summary: {len(expenses)} expenses totaling ${total:.2f}")
````

````python
# Example: Collect expenses until user says "done"
print("=== Multiple Expense Entry ===")

expenses = []
total = 0

print("Enter your expenses (type 'done' when finished):")

while True:
    expense_input = input("Expense amount (or 'done'): $")

    if expense_input.lower() == 'done':
        break

    try:
        expense = float(expense_input)
        if expense >= 0:
            expenses.append(expense)
            total += expense
            print(f"  Added ${expense:.2f} | Running total: ${total:.2f}")
        else:
            print("  Expense must be positive!")
    except ValueError:
        print("  Please enter a valid number or 'done'")

print(f"\nSummary: {len(expenses)} expenses totaling ${total:.2f}")

# ===== YOUR CHALLENGE: CATEGORY ENTRY =====
print("\n=== CELESTI Expense Categories Setup ===")

categories = []

print("Enter expense categories (type 'finish' when done):")

while True:
    category = input("Enter category name (or 'finish'): ")

    if category.lower() == 'finish':
        break

    if category.strip() == "":
        print("CELESTI says: Category cannot be empty!")
        continue

    categories.append(category)
    print(f"CELESTI saved: '{category}' | Total categories: {len(categories)}")

print(f"\nCELESTI saved {len(categories)} categories: {categories}")
````

## Part 4: For Loops with Known Counts (10 minutes)

### 📅 Weekly Expense Planning

Sometimes you know exactly how many items you need. Perfect for `for` loops!

````python
# Example: Get expenses for each day of the work week
print("=== Weekly Expense Planner ===")

days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
weekly_expenses = []
total_week = 0

print("Plan your expenses for each workday:")

for day in days:
    # Get planned expense for this day
    while True:  # Validation loop inside for loop!
        try:
            expense_str = input(f"{day} planned expense: $")
            expense = float(expense_str)
            if expense >= 0:
                weekly_expenses.append(expense)
                total_week += expense
                break  # Valid input, exit validation loop
            else:
                print("  Expense must be non-negative")
        except ValueError:
            print("  Please enter a valid number")

print(f"\n📅 Weekly Plan: ${total_week:.2f} total")
print(f"📊 Daily average: ${total_week/len(days):.2f}")
````

````python
# Example: Get expenses for each day of the work week
print("=== Weekly Expense Planner ===")

days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
weekly_expenses = []
total_week = 0

print("Plan your expenses for each workday:")

for day in days:
    while True:
        try:
            expense_str = input(f"{day} planned expense: $")
            expense = float(expense_str)
            if expense >= 0:
                weekly_expenses.append(expense)
                total_week += expense
                break
            else:
                print("  Expense must be non-negative")
        except ValueError:
            print("  Please enter a valid number")

print(f"\nWeekly Plan: ${total_week:.2f} total")
print(f"Daily average: ${total_week/len(days):.2f}")

# ===== YOUR CHALLENGE: SAVINGS GOALS =====
print("\n=== CELESTI Savings Goals Tracker ===")

goals = []

print("Let's set up 3 savings goals:")

for i in range(3):
    print(f"\n--- CELESTI Goal {i+1} ---")

    goal_name = input("Goal name: ")

    target = 0
    while target <= 0:
        try:
            target = float(input("Target amount: $"))
            if target <= 0:
                print("CELESTI says: Target must be greater than $0!")
        except ValueError:
            print("CELESTI says: Please enter a valid number!")

    current = -1
    while current < 0 or current > target:
        try:
            current = float(input("Current amount saved: $"))
            if current < 0:
                print("CELESTI says: Amount cannot be negative!")
            elif current > target:
                print(f"CELESTI says: Cannot exceed target of ${target:.2f}!")
        except ValueError:
            print("CELESTI says: Please enter a valid number!")

    percent = (current / target) * 100
    goals.append({"name": goal_name, "target": target, "current": current})
    print(f"CELESTI Progress: ${current:.2f} / ${target:.2f} ({percent:.1f}% complete)")

print("\n--- CELESTI Goals Summary ---")
for goal in goals:
    percent = (goal["current"] / goal["target"]) * 100
    print(f"  {goal['name']}: ${goal['current']:.2f} / ${goal['target']:.2f} ({percent:.1f}% complete)")
````

````python
print("=== CELESTI Finance Tracker v0.2 ===")
print("Week 3: Now with Loop Power!")
print()

# SETUP PHASE
name = input("Enter your name: ")

budget = 0
while budget < 100 or budget > 50000:
    try:
        budget_str = input("Enter monthly budget ($100-$50,000): $")
        budget = float(budget_str)
        if budget < 100:
            print("CELESTI says: Budget too low! Minimum $100")
        elif budget > 50000:
            print("CELESTI says: Budget too high! Maximum $50,000")
    except ValueError:
        print("CELESTI says: Please enter a valid number!")
        budget = 0

print(f"\nWelcome {name}! Budget: ${budget:.2f}")

# TRANSACTION COLLECTION
expenses = []
descriptions = []
total_spent = 0

print("\n--- Enter Your Expenses ---")
print("Type 'done' when finished")

while True:
    expense_input = input("\n[CELESTI] Expense amount (or 'done'): $")

    if expense_input.lower() == 'done':
        break

    try:
        amount = float(expense_input)
        if amount <= 0:
            print("CELESTI says: Please enter a positive amount!")
            continue

        description = input("[CELESTI] What was this expense for? ")
        expenses.append(amount)
        descriptions.append(description)
        total_spent += amount

        budget_percent = (total_spent / budget) * 100
        print(f"CELESTI recorded: ${amount:.2f} for '{description}'")
        print(f"Running total: ${total_spent:.2f} ({budget_percent:.1f}% of budget)")

        if budget_percent >= 100:
            print("CELESTI Alert: You have exceeded your budget!")
        elif budget_percent >= 80:
            print("CELESTI Warning: You are near your budget limit!")

    except ValueError:
        print("CELESTI says: Invalid amount! Enter a number or 'done'.")

# ANALYSIS PHASE
print(f"\nCELESTI Expense Analysis:")
print(f"Total expenses: {len(expenses)}")
print(f"Total spent: ${total_spent:.2f}")
print(f"Budget used: {(total_spent/budget)*100:.1f}%")
print(f"Remaining: ${budget - total_spent:.2f}")

large_expenses = 0
for amount in expenses:
    if (amount / budget) * 100 > 10:
        large_expenses += 1

print(f"Large expenses (>10% budget): {large_expenses}")

# MENU SYSTEM
print("\n=== CELESTI Finance Menu ===")

running = True
while running:
    print("\n" + "-"*30)
    print(f"CELESTI Finance Tracker - {name}")
    if len(expenses) > 0:
        print(f"Current: {len(expenses)} expenses, ${sum(expenses):.2f} spent")
    print("-"*30)
    print("1. View expense summary")
    print("2. Add new expense")
    print("3. Budget check")
    print("4. Exit")

    choice = input("\nChoose option (1-4): ")

    if choice == "1":
        print("\n--- CELESTI Expense Summary ---")
        if len(expenses) == 0:
            print("CELESTI: No expenses recorded yet!")
        else:
            for i in range(len(expenses)):
                print(f"  {i+1}. ${expenses[i]:.2f} - {descriptions[i]}")
            print(f"\n  Total: ${sum(expenses):.2f}")

    elif choice == "2":
        print("\n--- CELESTI Add New Expense ---")
        new_amount = 0
        while new_amount <= 0:
            try:
                new_amount = float(input("Expense amount: $"))
                if new_amount <= 0:
                    print("CELESTI says: Please enter a positive amount!")
            except ValueError:
                print("CELESTI says: Please enter a valid number!")

        new_description = input("What was this expense for? ")
        expenses.append(new_amount)
        descriptions.append(new_description)
        total_spent += new_amount

        budget_percent = (sum(expenses) / budget) * 100
        print(f"CELESTI recorded: ${new_amount:.2f} for '{new_description}'")
        print(f"New total: ${sum(expenses):.2f} ({budget_percent:.1f}% of budget)")

    elif choice == "3":
        print("\n--- CELESTI Budget Check ---")
        spent = sum(expenses) if expenses else 0
        budget_percent = (spent / budget) * 100
        remaining = budget - spent
        daily_budget = remaining / 30

        print(f"Budget:       ${budget:.2f}")
        print(f"Spent:        ${spent:.2f} ({budget_percent:.1f}%)")
        print(f"Remaining:    ${remaining:.2f}")
        print(f"Daily budget: ${daily_budget:.2f}/day for next 30 days")

        if budget_percent >= 100:
            print("CELESTI Advice: You are over budget! Stop spending now.")
        elif budget_percent >= 80:
            print("CELESTI Advice: Almost at your limit! Be very careful.")
        elif budget_percent >= 50:
            print("CELESTI Advice: Halfway through budget. Stay on track!")
        else:
            print("CELESTI Advice: You are in great shape. Keep it up!")

    elif choice == "4":
        running = False
        print(f"\nGoodbye {name}, thank you for using CELESTI!")
        print(f"Final summary: {len(expenses)} expenses, ${sum(expenses) if expenses else 0:.2f} total")
        print("CELESTI: Keep saving and growing stronger every week!")

    else:
        print("CELESTI says: Invalid choice! Please choose 1-4.")

print("\n=== CELESTI Finance Tracker Session Complete ===")
````
