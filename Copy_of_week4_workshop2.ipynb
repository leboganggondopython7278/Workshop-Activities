## Part 1: From Variables to Collections (8 minutes)

### 🧠 The Collection Problem

Your Workshop 1 tracker can handle multiple expenses, but it can't remember them for analysis. Let's fix that!

````python
# Problem: Workshop 1 approach - data gets lost
print("=== Workshop 1 Limitation Demo ===")

total = 0
count = 0

for i in range(3):
    expense = (i + 1) * 25
    total += expense
    count += 1
    print(f"Expense {i+1}: ${expense}")

print(f"\nTotal: ${total}, Count: {count}")
print("Problem: Individual expenses are LOST!")
print("Can't find largest expense")
print("Can't categorize by amount")
print("Can't search transaction history")

# Solution: Store in collections for later analysis
print("\n=== Collection-Powered Solution ===")

expenses = []
descriptions = []
total = 0

for i in range(3):
    expense = (i + 1) * 25
    description = f"Item {i+1}"
    expenses.append(expense)
    descriptions.append(description)
    total += expense

print(f"All expenses: {expenses}")
print(f"All descriptions: {descriptions}")
print("\nNow we can analyze patterns!")
print(f"Largest expense: ${max(expenses)}")
print(f"Smallest expense: ${min(expenses)}")
print(f"Average expense: ${total/len(expenses):.2f}")

expensive_count = 0
for expense in expenses:
    if expense > 40:
        expensive_count += 1

print(f"Expensive items (>$40): {expensive_count}")

# Your analogy completion:
# Lists are like a notebook with multiple pages because
# you can store many items and go back to read any of them later.

# Why does this matter for your finance tracker?
# Because we need to remember ALL expenses not just the last one
# so we can analyze patterns, search history and calculate totals.
````

````python
# Visual demonstration of indexing
monthly_expenses = [450.00, 380.50, 520.25, 290.75]

print("Your monthly expenses:")
print(f"List: {monthly_expenses}")
print(f"Index positions: 0, 1, 2, 3")
print()

print("Accessing by index:")
print(f"First month (index 0): ${monthly_expenses[0]}")
print(f"Second month (index 1): ${monthly_expenses[1]}")
print(f"Last month (index -1): ${monthly_expenses[-1]}")
print(f"Total months: {len(monthly_expenses)}")

print("\nCommon mistake:")
try:
    print(f"Month 4 (index 4): ${monthly_expenses[4]}")
except IndexError as e:
    print(f"IndexError: {e}")
    print("Remember: 4 items have indices 0,1,2,3 (not 1,2,3,4)")

# Safe Transaction Display
transactions = []
transactions.append(25.50)
transactions.append(67.25)
transactions.append(15.00)

print("\n=== Safe Transaction Display ===")

if len(transactions) > 0:
    print(f"First transaction: ${transactions[0]}")
    print(f"Latest transaction: ${transactions[-1]}")
    print(f"Total transactions: {len(transactions)}")

    print("\nAll transactions:")
    for i in range(len(transactions)):
        print(f"  {i+1}. ${transactions[i]:.2f}")

else:
    print("No transactions recorded yet")

empty_list = []
print(f"\nEmpty list test: {len(empty_list)} items")
if len(empty_list) > 0:
    print(f"First: {empty_list[0]}")
else:
    print("Safely handled empty list!")
````

## Part 3: Collection Processing Patterns (12 minutes)

### 📊 Financial Data Analysis Patterns

Now that you can store transaction data, let's analyze it! These patterns work for ANY collection analysis.

````python
# Sample transaction data
expenses = [23.50, 156.75, 45.00, 89.25, 12.80, 234.60, 67.30]
categories = ["Food", "Rent", "Food", "Gas", "Food", "Rent", "Entertainment"]

print("=== Financial Analysis Patterns ===")
print(f"Sample data: {len(expenses)} transactions")
print()

# Pattern 1: Accumulation
total = 0
for expense in expenses:
    total += expense
average = total / len(expenses)
print(f"Total spent: ${total:.2f}")
print(f"Average per transaction: ${average:.2f}")

# Pattern 2: Conditional Counting
large_expenses = 0
large_threshold = 100
for expense in expenses:
    if expense > large_threshold:
        large_expenses += 1
print(f"Large expenses (>${large_threshold}): {large_expenses}")

# Pattern 3: Finding Extremes
if len(expenses) > 0:
    highest = expenses[0]
    lowest = expenses[0]
    for expense in expenses:
        if expense > highest:
            highest = expense
        if expense < lowest:
            lowest = expense
    print(f"Highest expense: ${highest}")
    print(f"Lowest expense: ${lowest}")
    print(f"Spending range: ${highest - lowest:.2f}")

# Category Analysis Challenge
print("\n=== Category Analysis Challenge ===")

food_count = 0
rent_count = 0
gas_count = 0
entertainment_count = 0
food_total = 0
rent_total = 0
gas_total = 0
entertainment_total = 0

for i in range(len(categories)):
    if categories[i] == "Food":
        food_count += 1
        food_total += expenses[i]
    elif categories[i] == "Rent":
        rent_count += 1
        rent_total += expenses[i]
    elif categories[i] == "Gas":
        gas_count += 1
        gas_total += expenses[i]
    elif categories[i] == "Entertainment":
        entertainment_count += 1
        entertainment_total += expenses[i]

print("\nCategory Summary:")
print(f"Food: {food_count} transactions, ${food_total:.2f} total")
print(f"Rent: {rent_count} transactions, ${rent_total:.2f} total")
print(f"Gas: {gas_count} transactions, ${gas_total:.2f} total")
print(f"Entertainment: {entertainment_count} transactions, ${entertainment_total:.2f} total")
````

## Part 2: Search and Filter Patterns (10 minutes)

### 🔍 Finding Specific Transactions

Real finance trackers need search capabilities. Let's build the fundamental patterns:

### 🎯 Your Challenge: Amount-Based Filtering

Build a filter that finds all transactions within a specific amount range:

````python
# Sample transaction database
amounts = [23.50, 156.75, 45.00, 89.25, 12.80, 234.60, 67.30]
descriptions = ["Grocery", "Rent", "Restaurant", "Gas", "Coffee", "Rent", "Movie"]
dates = ["Jan 1", "Jan 1", "Jan 2", "Jan 3", "Jan 3", "Feb 1", "Feb 5"]

print("=== Transaction Search Demo ===")

# Search by description
search_term = "Rent"
found_transactions = []

print(f"Searching for transactions containing '{search_term}':")

for i in range(len(descriptions)):
    if search_term.lower() in descriptions[i].lower():
        transaction_info = {
            'amount': amounts[i],
            'description': descriptions[i],
            'date': dates[i]
        }
        found_transactions.append(transaction_info)
        print(f"  {dates[i]}: ${amounts[i]} - {descriptions[i]}")

print(f"\nFound {len(found_transactions)} matching transactions")

search_total = 0
for transaction in found_transactions:
    search_total += transaction['amount']
print(f"Total for '{search_term}': ${search_total:.2f}")
````

````python
# Amount Range Filter
amounts = [23.50, 156.75, 45.00, 89.25, 12.80, 234.60, 67.30]
descriptions = ["Grocery", "Rent", "Restaurant", "Gas", "Coffee", "Rent", "Movie"]

print("=== Amount Range Filter ===")

min_amount = float(input("Minimum amount: $"))
max_amount = float(input("Maximum amount: $"))

print(f"\nFinding transactions between ${min_amount} and ${max_amount}:")

matching_count = 0
matching_total = 0

for i in range(len(amounts)):
    if amounts[i] >= min_amount and amounts[i] <= max_amount:
        print(f"  {descriptions[i]}: ${amounts[i]:.2f}")
        matching_count += 1
        matching_total += amounts[i]

print(f"\nFilter Results:")
print(f"Matching transactions: {matching_count}")
print(f"Total in range: ${matching_total:.2f}")

# Advanced Search
dates = ["Jan 1", "Jan 1", "Jan 2", "Jan 3", "Jan 3", "Feb 1", "Feb 5"]

print("\n=== Advanced Search: January Expenses Over $20 ===")

for i in range(len(dates)):
    if "Jan" in dates[i] and amounts[i] > 20:
        print(f"  {dates[i]}: ${amounts[i]:.2f} - {descriptions[i]}")
````

## Part 3: Interactive Menu Systems (15 minutes)

### 🎮 Building User-Controlled Programs

Combine everything you've learned into an interactive finance tracker that keeps running until the user decides to quit:

````python
print("=== CELESTI Interactive Finance Tracker v0.3 ===")
print("Workshop 2: Lists + Loops = Power!")
print()

name = input("Enter your name: ")
expenses = []
descriptions = []
budget = 0

while budget < 100 or budget > 50000:
    try:
        budget_str = input("Monthly budget ($100-$50,000): $")
        budget = float(budget_str)
        if budget < 100:
            print("CELESTI says: Budget too low! Minimum $100")
        elif budget > 50000:
            print("CELESTI says: Budget too high! Maximum $50,000")
    except ValueError:
        print("CELESTI says: Please enter a valid number!")
        budget = 0

print(f"\nWelcome {name}! Budget: ${budget:.2f}")

running = True
while running:
    print("\n" + "="*40)
    print(f"CELESTI Finance Tracker - {name}")
    print(f"Current: {len(expenses)} transactions, ${sum(expenses):.2f} spent")
    print("="*40)
    print("1. Add expense")
    print("2. View all transactions")
    print("3. Search transactions")
    print("4. Budget analysis")
    print("5. Exit")

    choice = input("\nChoose option (1-5): ").strip()

    if choice == "1":
        print("\n--- Add Expense ---")
        new_amount = 0
        while new_amount <= 0:
            try:
                new_amount = float(input("Expense amount: $"))
                if new_amount <= 0:
                    print("CELESTI says: Please enter a positive amount!")
            except ValueError:
                print("CELESTI says: Please enter a valid number!")
                new_amount = 0
        new_description = input("What was this expense for? ")
        expenses.append(new_amount)
        descriptions.append(new_description)
        budget_percent = (sum(expenses) / budget) * 100
        print(f"CELESTI recorded: ${new_amount:.2f} for '{new_description}'")
        print(f"Running total: ${sum(expenses):.2f} ({budget_percent:.1f}% of budget)")
        if budget_percent >= 100:
            print("CELESTI Alert: You have exceeded your budget!")
        elif budget_percent >= 80:
            print("CELESTI Warning: You are near your budget limit!")

    elif choice == "2":
        print("\n--- CELESTI Transaction History ---")
        if len(expenses) == 0:
            print("CELESTI: No transactions yet!")
        else:
            for i in range(len(expenses)):
                print(f"  {i+1}. ${expenses[i]:.2f} - {descriptions[i]}")
            print(f"\n  Total: ${sum(expenses):.2f}")

    elif choice == "3":
        print("\n--- CELESTI Search Transactions ---")
        if len(expenses) == 0:
            print("CELESTI: No transactions to search!")
        else:
            search_term = input("Search for: ")
            found = False
            search_total = 0
            for i in range(len(descriptions)):
                if search_term.lower() in descriptions[i].lower():
                    print(f"  ${expenses[i]:.2f} - {descriptions[i]}")
                    search_total += expenses[i]
                    found = True
            if found:
                print(f"  Total for '{search_term}': ${search_total:.2f}")
            else:
                print(f"  No transactions found for '{search_term}'")

    elif choice == "4":
        print("\n--- CELESTI Budget Analysis ---")
        if len(expenses) == 0:
            print("CELESTI: No transactions to analyze!")
        else:
            total = sum(expenses)
            average = total / len(expenses)
            budget_percent = (total / budget) * 100
            remaining = budget - total

            highest = expenses[0]
            lowest = expenses[0]
            for e in expenses:
                if e > highest:
                    highest = e
                if e < lowest:
                    lowest = e

            large_count = 0
            for e in expenses:
                if (e / budget) * 100 > 10:
                    large_count += 1

            print(f"Total spent:     ${total:.2f}")
            print(f"Average:         ${average:.2f}")
            print(f"Budget used:     {budget_percent:.1f}%")
            print(f"Remaining:       ${remaining:.2f}")
            print(f"Highest expense: ${highest:.2f}")
            print(f"Lowest expense:  ${lowest:.2f}")
            print(f"Large expenses:  {large_count}")

            if budget_percent >= 100:
                print("CELESTI Advice: You are over budget! Stop spending now.")
            elif budget_percent >= 80:
                print("CELESTI Advice: Almost at limit! Be very careful.")
            elif budget_percent >= 50:
                print("CELESTI Advice: Halfway through. Stay on track!")
            else:
                print("CELESTI Advice: You are in great shape. Keep it up!")

    elif choice == "5":
        running = False
        print(f"\nGoodbye {name}, thank you for using CELESTI!")
        print(f"Final: {len(expenses)} expenses, ${sum(expenses):.2f} total")
        print("CELESTI: Keep saving and growing stronger!")

    else:
        print("CELESTI says: Invalid choice! Please choose 1-5.")

print("\n=== CELESTI Finance Tracker Session Complete ===")
````

## Part 4: Advanced Pattern Application (8 minutes)

### 📈 Spending Trend Analysis

Let's build more sophisticated analysis using the patterns you've mastered:

````python
# Spending Trend Analysis
print("=== CELESTI Spending Pattern Analyzer ===")

weekly_totals = [156.75, 234.50, 198.25, 267.80, 145.60]
week_names = ["Week 1", "Week 2", "Week 3", "Week 4", "Week 5"]

print("Weekly spending analysis:")

for i in range(len(weekly_totals)):
    current_week = weekly_totals[i]

    if i > 0:
        previous_week = weekly_totals[i-1]
        change_percent = ((current_week - previous_week) / previous_week) * 100

        if current_week > previous_week:
            trend = "Up"
        elif current_week < previous_week:
            trend = "Down"
        else:
            trend = "Same"

        print(f"{week_names[i]}: ${current_week:.2f} [{trend} {abs(change_percent):.1f}%]")
    else:
        print(f"{week_names[i]}: ${current_week:.2f} [Baseline]")

# Best and worst weeks
highest_week = weekly_totals[0]
lowest_week = weekly_totals[0]
highest_name = week_names[0]
lowest_name = week_names[0]

for i in range(len(weekly_totals)):
    if weekly_totals[i] > highest_week:
        highest_week = weekly_totals[i]
        highest_name = week_names[i]
    if weekly_totals[i] < lowest_week:
        lowest_week = weekly_totals[i]
        lowest_name = week_names[i]

print(f"\nBest week (lowest spending):  {lowest_name} ${lowest_week:.2f}")
print(f"Worst week (highest spending): {highest_name} ${highest_week:.2f}")

first_week = weekly_totals[0]
last_week = weekly_totals[-1]
if last_week < first_week:
    print("Overall trend: Spending is going DOWN - great progress!")
elif last_week > first_week:
    print("Overall trend: Spending is going UP - be careful!")
else:
    print("Overall trend: Spending is STABLE")
````
